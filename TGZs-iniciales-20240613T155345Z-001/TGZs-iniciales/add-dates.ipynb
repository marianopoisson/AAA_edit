```python
topdate='09 February 2024'
art=[]

# Función para actualizar la línea "accepted:" en un archivo .tex
def update_tex_file(tex_file, date):
    with open(tex_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    with open(tex_file, 'w', encoding='utf-8') as file:
        for line in lines:
            if '\\received{' in line:
                print(tex_file)
                file.write('\\received{'+ topdate+'}\n')
            elif '\\accepted{' in line:
                file.write('\\accepted{'+ date+'}\n')
            else:
                file.write(line)
    
# Función para procesar un archivo .tgz
def process_tgz(tgz_file, date):
    with tarfile.open(tgz_file, 'r:gz') as tar:
        extracted_dir = os.path.splitext(os.path.basename(tgz_file))[0]
        tar.extractall(path=extracted_dir)
        tex_file = None

        for root, dirs, files in os.walk(extracted_dir):
            for file in files:
                if file.endswith('.tex'):
                    tex_file = os.path.join(root, file)
                    update_tex_file(tex_file, date)
                    break

    # Crear un nuevo archivo .tgz con el .tex actualizado
    new_tgz_file = f'{extracted_dir}_updated.tgz'
    with tarfile.open(new_tgz_file, 'w:gz') as tar:
        tar.add(extracted_dir, arcname=os.path.basename(extracted_dir))

    # Limpiar la carpeta extraída
    os.system(f'rm -rf {extracted_dir}')

# Leer las fechas del archivo CSV
dates = {}
with open('fechas.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        file_id,titulo,area,estado, date = row
        dates[file_id] = date.replace('may','May').replace('mar','March').replace('abr','April').replace('aug','August').replace('ago','August').replace('jun','June').replace('jul','July')
# Procesar cada archivo .tgz en la carpeta actual

os.chdir('./tgz2/')
for file in os.listdir('.'):
    if file.endswith('.tgz'):
        file_id = os.path.splitext(file)[0]
        if file_id in dates:
            process_tgz(file, dates[file_id])
os.chdir('..')
print('Proceso completado.')
```
