```python
import os
import re
import shutil
from collections import defaultdict

# Define los directorios
directorio_origen = 'tex'
directorio_destino = 'tex1'

# Crear el directorio de salida si no existe
os.makedirs(directorio_destino, exist_ok=True)

# Patrón de nombre de archivo: ####_v#.tex
patron = re.compile(r'^(\d{4})_v(\d+)\.tex$')

# Diccionario para agrupar versiones por número base
archivos_por_numero = defaultdict(list)

# Recorre los archivos del directorio de entrada
for archivo in os.listdir(directorio_origen):
    match = patron.match(archivo)
    if match:
        numero = match.group(1)
        version = int(match.group(2))
        archivos_por_numero[numero].append((version, archivo))

# Procesa cada grupo
for numero, versiones in archivos_por_numero.items():
    # Encuentra el de mayor versión
    version_maxima, archivo_maximo = max(versiones, key=lambda x: x[0])
    nuevo_nombre = f"{numero}.tex"
    
    ruta_origen = os.path.join(directorio_origen, archivo_maximo)
    ruta_destino = os.path.join(directorio_destino, nuevo_nombre)

    # Copia y renombra
    shutil.copy(ruta_origen, ruta_destino)
    print(f"✅ Copiado: {archivo_maximo} → {nuevo_nombre}")
```
