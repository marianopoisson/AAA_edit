```python
import os
import re

```

```python
# Directorio donde se encuentran los archivos
directorio = "./corregidos"

# Iterar sobre los archivos en el directorio
for nombre_archivo in os.listdir(directorio):
    # Buscar archivos que sigan el patrón ####_v*.tex
    coincidencia = re.match(r'(\d{4})_v\d+corr.tex', nombre_archivo)
    
    if coincidencia:
        # Extraer el número ####
        nuevo_nombre = coincidencia.group(1) + ".tex"
        ruta_vieja = os.path.join(directorio, nombre_archivo)
        ruta_nueva = os.path.join(directorio, nuevo_nombre)
        
        # Renombrar el archivo
        os.rename(ruta_vieja, ruta_nueva)
        print(f"Renombrado: {nombre_archivo} -> {nuevo_nombre}")

print("Proceso completado.")
```

```python
import os
import shutil

# Directorios de origen y destino
directorio_origen = "./tex"
directorio_destino = "./tex2"

# Ruta del archivo lista.txt
ruta_lista = "listamurales.txt"

# Leer los números de archivos desde lista.txt
with open(ruta_lista, 'r') as archivo_lista:
    numeros_a_copiar = [linea.strip() for linea in archivo_lista.readlines()]

# Copiar los archivos listados en lista.txt
for numero in numeros_a_copiar:
    nombre_archivo = f"{numero}.tex"
    ruta_origen = os.path.join(directorio_origen, nombre_archivo)
    ruta_destino = os.path.join(directorio_destino, nombre_archivo)
    
    # Verificar si el archivo existe antes de copiar
    if os.path.exists(ruta_origen):
        # Copiar el archivo
        shutil.copy(ruta_origen, ruta_destino)
        print(f"Copiado: {nombre_archivo} a {ruta_destino}")
    else:
        print(f"El archivo {nombre_archivo} no existe en el directorio de origen.")

print("Proceso completado.")
```

```python
# Directorio donde se encuentran los archivos
directorio = "./corregidos"

# Diccionario para almacenar la versión más alta de cada número base ####
archivos_por_numero = {}

# Iterar sobre los archivos en el directorio
for nombre_archivo in os.listdir(directorio):
    # Buscar archivos que sigan el patrón ####_v*.tex
    coincidencia = re.match(r'(\d{4})_v(\d+)\.tex', nombre_archivo)
    
    if coincidencia:
        numero_base = coincidencia.group(1)  # ####
        version = int(coincidencia.group(2))  # versión numérica v*

        # Si no existe un archivo para ese número o si la versión es mayor, actualizamos
        if numero_base not in archivos_por_numero or archivos_por_numero[numero_base][1] < version:
            archivos_por_numero[numero_base] = (nombre_archivo, version)

# Renombrar solo los archivos con la versión más alta
for numero_base, (nombre_archivo, version) in archivos_por_numero.items():
    nuevo_nombre = numero_base + ".tex"
    ruta_vieja = os.path.join(directorio, nombre_archivo)
    ruta_nueva = os.path.join(directorio, nuevo_nombre)
    
    # Renombrar el archivo
    os.rename(ruta_vieja, ruta_nueva)
    print(f"Renombrado: {nombre_archivo} -> {nuevo_nombre}")

print("Proceso completado.")
```

```python
import os

# Directorio donde se encuentran los archivos .tex
directorio = "./tex2"

# Ruta del archivo lista.txt que contiene los nombres de los archivos .tex
ruta_lista = "./listaorales.txt"

# Leer los números de archivos desde lista.txt
with open(ruta_lista, 'r') as archivo_lista:
    archivos_a_revisar = [linea.strip() + ".tex" for linea in archivo_lista.readlines()]

# Iterar sobre los archivos listados
for nombre_archivo in archivos_a_revisar:
    ruta_archivo = os.path.join(directorio, nombre_archivo)
    
    # Verificar si el archivo existe
    if os.path.exists(ruta_archivo):
        # Leer el contenido del archivo
        with open(ruta_archivo, 'r') as archivo:
            contenido = archivo.read()

        # Revisar si contiene \contribtype{1} y reemplazarlo por \contribtype{2}
        if r"\contribtype{1}" in contenido:
            contenido_nuevo = contenido.replace(r"\contribtype{1}", r"\contribtype{2}")
            
            # Guardar los cambios en el archivo
            with open(ruta_archivo, 'w') as archivo:
                archivo.write(contenido_nuevo)
            
            print(f"Modificado: {nombre_archivo} - \contribtype{1} cambiado a \contribtype{2}")
        else:
            print(f"Sin cambios: {nombre_archivo} - No contiene \contribtype{{1}}")
    else:
        print(f"El archivo {nombre_archivo} no existe.")

print("Proceso completado.")
```

```python
import os

# Directorio donde se encuentran los archivos .tex
directorio = "./tex2"

# Ruta del archivo lista.txt que contiene los nombres de los archivos .tex
ruta_lista = "./listamurales.txt"

# Leer los números de archivos desde lista.txt
with open(ruta_lista, 'r', encoding='utf-8') as archivo_lista:
    archivos_a_revisar = [linea.strip() + ".tex" for linea in archivo_lista.readlines()]

# Iterar sobre los archivos listados
for nombre_archivo in archivos_a_revisar:
    ruta_archivo = os.path.join(directorio, nombre_archivo)
    
    # Verificar si el archivo existe
    if os.path.exists(ruta_archivo):
        try:
            # Leer el contenido del archivo usando utf-8
            with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
                contenido = archivo.read()

            # Revisar si contiene \contribtype{1} y reemplazarlo por \contribtype{2}
            if r"\contribtype{2}" in contenido:
                contenido_nuevo = contenido.replace(r"\contribtype{2}", r"\contribtype{1}")
                
                # Guardar los cambios en el archivo usando utf-8
                with open(ruta_archivo, 'w', encoding='utf-8') as archivo:
                    archivo.write(contenido_nuevo)
                
                print(f"Modificado: {nombre_archivo} - \contribtype{{2}} cambiado a \contribtype{{1}}")
            else:
                print(f"Sin cambios: {nombre_archivo} - No contiene \contribtype{{2}}")
        except UnicodeDecodeError as e:
            print(f"Error de codificación al leer {nombre_archivo}: {e}")
    else:
        print(f"El archivo {nombre_archivo} no existe.")

print("Proceso completado.")
```
